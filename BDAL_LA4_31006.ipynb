```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('boston_housing.csv - boston_housing.csv.csv')
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
x=df.iloc[:,:]
y=df.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
y_predicted=model.predict(x_test)
y_predicted
```

```python
df=pd.DataFrame({'real values':y_test,'predicted values':y_predicted})
df
```

```python
#Mean Squared Error (MSE): 
#Calculates the average squared difference between predicted and actual values. Lower MSE indicates better performance.
mse=mean_squared_error(y_test,y_predicted)
mse
```

```python
print(np.mean((y_predicted-y_test)**2))
```

```python
plt.scatter(y_test,y_predicted)
plt.xlabel('True values')
plt.ylabel('Predicted')
plt.plot()
plt.show()
```
